Hypothesis: Increase in service hours in a given week reflects an increase
in new clients from the previous week.

First,read the dataset from local drive

```python
import pandas as pd
```

```python
df = pd.read_excel('New Clients and Service Hours.xlsx')[['Week ending','New Client','Service']]
df
```

Our goal is to determine the pattern of the increase of service  rather than the service itself,so I added the following code to add the difference of service compared to the last period

```python
change = df['Service'].diff()
```

```python
df['change'] = change
```

```python
df
```

Based on my understanding...increase should not be negative, otherwise it should be stated as 'change' or 'difference', so I only kept the positive data points for my hypothesis testing...

```python
test = df[(df['change']>0)&(df['New Client']>0)].copy()
test
```

I wanted to see the plot between these two variables first..

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.scatter(test['New Client'],test['change'])
plt.show()
```

It does not show a strong pattern, but it does not mean they are independent...perform a hypothesis testing anyway


Here I thought a chi-square independence test would be appropriate, as we were to determine the independence of these variables if we simply paraphrase the original hypothesis above.

The null hypothesis becomes: Increase in service hours in a given week and increase
in new clients from the previous week are independent

alternative: they are not independent.

import existing package and plug in the data...

```python
from scipy.stats import chi2_contingency
```

```python
g, p_value, dof, expctd = chi2_contingency(test[['New Client','change']])
```

```python
p_value
```

Under 0.05 significance level, we reject the null hypothesis and conclude that there's dependence between these variables

Conclusion:
Increase in service hours in a given week  does reflect an increase
in new clients from the previous week.
